```python
data={'A':[],                   # cretaing empty dict
      'B':[],
      'C':[],
      'D':[]}

for i in range(1,500,5):        # fill key A
    data['A'].append(i)
for i in range(1,700,7):        # fill key B   
    data['B'].append(i)
for i in range(1,800,8):        # fill key C
    data['C'].append(i)
for i in range(1,1800,18):       # fill key C
    data['D'].append(i)
# shuffleing data    
# r.shuffle(data["A"])
r.shuffle(data["B"])
r.shuffle(data["C"])
r.shuffle(data["D"])  

print(len(data['A']) is len(data['B']),len(data['C']) is len(data['D']))          # checking that length matches
print(len(data['A']),len(data['B']))
```

# Line Plot

```python
sns.lineplot(df,y='bill_depth_mm',x='bill_length_mm',style='sex',dashes=[(5, 2),(1, 4)])
plt.show()
sns.lineplot(df,y='bill_depth_mm',x='bill_length_mm',style='sex', dashes={'Male':(5, 2),'Female':(1, 4)})
plt.show()
sns.lineplot(df,y='bill_depth_mm',x='bill_length_mm',errorbar='ci')
plt.show()
```

# Box Plot

```python
Df=pd.read_csv('tips.csv')
df=Df
print(df.head(2).to_string())
```

```python
sns.set_theme(style='whitegrid')
sns.boxplot(df,x='day',y='total_bill'
            ,hue='smoker',linewidth=2 ,boxprops={'facecolor':'orange'},width=1, palette=['red','orange'],saturation=1)
plt.show()
```

# Violin Plot

```python
sns.violinplot(df,x='day',y='tip',hue='sex',split=True)
plt.show()
```

```python
sns.violinplot(data=df,y='tip',density_norm='width')
plt.show()
```

```python
sns.violinplot(df,x='day',y='tip',hue='sex',scale='width',linewidth=1)
plt.show()
```

# Strip plot

```python
print(df.head(2))
```

```python
sns.stripplot(df,x='day',y='total_bill',hue='sex',jitter=True,dodge=True,linewidth=2)
plt.show()
```

```python
sns.scatterplot(df,y='tip',x='total_bill',hue='sex',style='sex',size='sex',sizes=(1,50))
plt.show()
```
