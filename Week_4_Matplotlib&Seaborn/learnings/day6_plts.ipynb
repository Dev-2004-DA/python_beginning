```python
import numpy as np , pandas as pd
import matplotlib.pyplot as plt , seaborn as sns
df=sns.load_dataset('anagrams')
```

# **Heat map**

# Conut plot

```python
df=sns.load_dataset('tips')
print(df.head(2))
sns.countplot(df , y='size')
plt.show()
```

```python
df = pd.DataFrame({
    'fruit': ['apple', 'banana', 'apple', 'apple', 'banana', 'mango']
})
# plt.figure(figsize=(8,2))
sns.countplot(data=df, x='fruit')
sns.despine(left=True)
plt.show()
```

# Pair plot

```python
df=sns.load_dataset('tips')
print(df.head(2))
sns.pairplot(df[['tip','total_bill']],corner=True)
# sns.pairplot(df,vars=['tip','total_bill'])
plt.show()
```

```python
sns.jointplot(df,x='sex',y='total_bill')
```

# Cat Plot

```python
df=df.head(20)
print(df.head( ))
# sns.catplot(df,x='tip',y='size',hue='sex',palette='plasma',  alpha=0.7,kind='bar',col='time',row='smoker')
sns.catplot(data=df, x='sex', y='tip', kind='bar', col='size', row='smoker',aspect=1.2)

plt.show()
```

# Histogram plot

```python
df=sns.load_dataset('penguins')
print(df.head(2).to_string())
sns.histplot(df,x='bill_length_mm',element='step',fill=False)
plt.show()
```

```python
# print(df.head(2).to_string())
# sns.kdeplot(df,x='bill_depth_mm',hue='island',common_norm=True)
# sns.kdeplot(df,x='bill_depth_mm',hue='island',common_norm=True,cut=12)
# sns.kdeplot(df,x='bill_depth_mm',hue='island',common_norm=True,gridsize=10)
sns.kdeplot(df,x='bill_depth_mm',hue='island',common_norm=True,clip=(14,20))
plt.show()
```
