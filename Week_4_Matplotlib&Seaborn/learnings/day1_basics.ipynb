```python
### formatiing chart aree
y=[2,3,4,5,1,2,3,0,6,8,7]
plt.plot(y)
plt.title('format',loc='left',fontdict={'family':'monospace','fontsize':22,'color':'magenta'})
plt.xlabel('X- Axis',fontdict={'family':'monospace','fontsize':18,'color':'blue'})
plt.ylabel('Y-axis',fontdict={'family':'monospace','fontsize':18,'color':'blue'})
plt.show()
```

```python
## Grid lines
plt.grid(axis='x')
plt.show()
```

```python
#Subplot
x= [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [4, 3, 2, 1]

plt.subplot(1, 2, 1)  # (rows=1, cols=2, this is 1st plot)
plt.plot(x, y1)
plt.title("Plot 1")

plt.subplot(1, 2, 2)  # (rows=1, cols=2, this is 2nd plot)
plt.plot(x, y2)
plt.title("Plot 2")
plt.tight_layout()
plt.show()
```

```python
#legend
plt.plot(x, y1,label='huhu', marker='d',ls='')
plt.legend(loc='right')
plt.show()
```
