```python
data = [60, 70, 80, 90]         # one box
# data = [[60, 70], [80, 90]] 
plt.boxplot(data,labels=['A'],patch_artist=True , notch=True)
plt.grid()
plt.show()
```

```python


data = [[60, 70, 75, 80], [80, 85, 88, 90]]

plt.boxplot(data, labels=['Group A', 'Group B'], notch=True, patch_artist=True)
plt.title("Boxplot with Notches")
plt.grid(True)
plt.show()
```

```python
fig,axs = plt.subplots(1, 2)  # 1 row, 2 subplots

axs[0].plot([1, 2, 3], [4, 5, 6])
axs[0].set_title("Left Plot")

axs[1].plot([1, 2, 3], [6, 5, 4])
axs[1].set_title("Right Plot")

# plt.subplot(1,2,1)
# plt.plot([1,2,3],[4,5,6])
# plt.title('Left')
# plt.subplot(1,2,2)
# plt.plot([1,2,3],[6,5,4])
# plt.title("Right")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

# First subplot
ax[0].hist([1, 2, 3], [4, 5, 6], label="Up")
ax[0].set_title("Increasing")
ax[0].legend()

# Second subplot
ax[1].plot([1, 2, 3], [6, 5, 4], label="Down")
ax[1].set_title("Decreasing")
ax[1].legend()

# Title for whole figure
fig.suptitle("Comparison of Trends")

plt.tight_layout()
plt.show()
```
