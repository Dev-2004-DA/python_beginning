```python
import matplotlib.pyplot as plt , numpy as np 
from numpy import random as r 
```

```python
#SCATTER PLOT
x=np.array([1,2,3,4,5,6,7])
y=np.array([2,3,4,4,3,2,1])
plt.scatter(x,y)
plt.show()
```

```python
# advanced color mapping
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()

sizes = np.random.randint(100, size=(100))
print(sizes)
```

```python
### plotting a customised scatter plot 
x=list(range(1,51,))
# y=r.randint(200,size=50)
z=y.copy()
print(z)
print(x)
print( len(z) is len(x))    ## checking that both length are equal or not 
color=[r.randint(100,size=50)]
plt.scatter(x,z,marker='D',c=color,cmap='plasma',alpha=0.7)
plt.title('Customized Scatter Plot')
plt.colorbar()
plt.show()
```

```python
## Pie chart
x=[1,2,3,5,6,8,9]
y=['A','B','C','D','E','F','G']
plt.pie(x,labels=y)
plt.show()
```

```python
#customization

#explode
x=[1,2,3,5,6,8,9]
y=['A','B','C','D','E','F','G']
plt.pie(x,labels=y,explode=[0.2,0.1,0.01,0.2,0.1,0.01,0])
plt.show()

#autopct
plt.pie(x,labels=y,autopct='%1.0f%%')
plt.show()
```
