```python
import matplotlib.pyplot as plt , numpy as np 
from numpy import random as r
```

```python
## bar charts

x=['A','B','C','D']
y=[2,5,8,3]
plt.subplot(1,2,1)
plt.bar(x,y, tick_label=['w','e','w','z'])
plt.subplot(1,2,2)
plt.barh(x,y)
plt.show()
```

```python
### histogram 
data = np.random.normal(50, 10, 200)  # 200 values ~ N(50,10)

plt.hist(data,
         bins=50,
         color='red',
         edgecolor='black',
         alpha=1)

plt.title("Histogram of Random Data")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.savefig('hist.pdf')
plt.show()
```
